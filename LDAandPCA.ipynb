```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

Iris

```python
iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
X.shape
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
```

```python
plt.figure()
colors = ["navy", "turquoise", "darkorange"]
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, lw=lw, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of IRIS dataset")
```

```python
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_r2[y == i, 0], X_r2[y == i, 1], alpha=0.8, color=color, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")
```

## **Digits**

```python
load_digits = datasets.load_digits();
```

```python
X = load_digits.data
y = load_digits.target
target_names = load_digits.target_names
```

```python
X.shape
```

```python
y.shape
```

```python
target_names
```

```python
pca = PCA(n_components=3)
X_r = pca.fit(X).transform(X)
```

```python
lda = LinearDiscriminantAnalysis(n_components=3)
X_r2 = lda.fit(X, y).transform(X)
```

```python
X_r2.shape
```

```python
colors_dict = mcolors.get_named_colors_mapping()
colors_list = [i[0] for i in colors_dict.items()]
rng = np.random.default_rng()
rints = rng.integers(low=0, high=len(colors_list)-1, size=target_names.shape[0])
colors = []
for rint in rints:
  colors.append(colors_list[rint])
```

```python
colors
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
lw = 3

for color, i, target_name in zip(colors, np.arange(0,target_names.shape[0]), target_names):
    ax.scatter(
        X_r[y == i, 0], X_r[y == i, 1], X_r[y == i, 2], color=color, alpha=0.8, lw=lw, label=target_name
    )
ax.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of Digits dataset")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
lw = 3
for color, i, target_name in zip(colors, np.arange(0,target_names.shape[0]), target_names):
    ax.scatter(
        X_r2[y == i, 0], X_r2[y == i, 1], X_r2[y == i, 2], alpha=0.8, color=color, label=target_name
    )
ax.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of Digits dataset")
```
